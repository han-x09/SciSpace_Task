```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import requests

url = 'https://drive.google.com/file/d/10Opb2w3Rb858fZqjg6cnkeTCOIENQUkC/view?usp=share_link'
response = requests.get(url)

from io import StringIO

data = StringIO(response.text)
df = pd.read_csv(data)

df =pd.read_csv('/Users/darshanpatel/Downloads/Crime_Reports.csv')
df
```

```python
df.drop(columns=['clearance_type','address_2','zip','country','state','city'],axis=1,inplace=True) 
```

```python
df.dropna()

df
```

```python
df['incident_datetime'] = pd.to_datetime(df['incident_datetime'])

# Getting year from incident_datetime
incident_years = df['incident_datetime'].dt.year

# Counting incidents in a year
incident_counts = incident_years.value_counts().sort_index()

# Plot the bar graph
plt.figure(figsize=(10, 6))
plt.bar(incident_counts.index, incident_counts.values)
plt.xlabel('Year')
plt.ylabel('Incident Count')
plt.title('Incidents by Year')
plt.xticks(incident_counts.index)
plt.show()
```

```python
df['incident_datetime'] = pd.to_datetime(df['incident_datetime'])

#  incidents in the year 2017
df_2017 = df[df['incident_datetime'].dt.year == 2017]

# month from the incident_datetime
incident_months = df_2017['incident_datetime'].dt.month

# counting incidents within each month
incident_counts = incident_months.value_counts().sort_index()
print(np.mean(incident_counts))

# Plot the bar graph
plt.figure(figsize=(10, 6))
plt.bar(incident_counts.index, incident_counts.values)
plt.xlabel('Month')
plt.ylabel('Incident Count')
plt.title('Incidents by Month in 2017')
plt.xticks(incident_counts.index)
plt.show()
```

```python
#What are the most common types of crimes in the city?

top_10_categories = df['parent_incident_type'].value_counts().head(10).index

filtered_df = df[df['parent_incident_type'].isin(top_10_categories)]

# Create a larger figure for the plot
plt.figure(figsize=(20, 10))

# Create a histogram
plt.hist(filtered_df['parent_incident_type'], bins=len(top_10_categories), edgecolor='black')
plt.xlabel('Incident Type')
plt.ylabel('Frequency')
plt.title('Histogram of Top 10 Types of Crime')
```

```python
df['incident_datetime'] = pd.to_datetime(df['incident_datetime'])

#  time interval is hour
time_intervals = df['incident_datetime'].dt.hour  

# counting incidents in a hour
incident_counts = time_intervals.value_counts().sort_index()

# Plot the bar graph
plt.figure(figsize=(10, 6))
plt.bar(incident_counts.index, incident_counts.values)
plt.xlabel('Hour of the Day')
plt.ylabel('Incident Count')
plt.title('Incidents by Hour of the Day')
plt.xticks(incident_counts.index)
plt.show()
```

```python
#b
incident_counts = df['day_of_week'].value_counts().reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])

# Plot the bar plot
plt.figure(figsize=(10, 6))
incident_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.xlabel('Day of the Week')
plt.ylabel('Incident Count')
plt.title('Incidents by Day of the Week')
plt.xticks(rotation=45)
plt.show()
```

```python
#Are there specific time-of-day patterns in crime occurrence?

pivot_table = df.pivot_table(index='hour_of_day', columns='parent_incident_type', aggfunc='size', fill_value=0)

# Create a heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(pivot_table, cmap='YlGnBu', annot=True, fmt='d', linewidths=0.5)
plt.xlabel('Parent Incident Type')
plt.ylabel('Hour of Day')
plt.title('Incidents by Hour of Day and Parent Incident Type')
plt.show()
```

```python
top_10 = df['address_1'].value_counts().head(10).index

filtered_df = df[df['address_1'].isin(top_10)]


plt.figure(figsize=(20, 10))

# Create a histogram
plt.hist(filtered_df['address_1'], bins=len(top_10), edgecolor='black')
plt.xlabel('Incident Type')
plt.ylabel('Frequency')
plt.title('Histogram of Top 10 Locations')
```

```python
crime_category = {
    'Assault': 'Violent',
    'Weapons Offense': 'Violent',
    'Other Sexual Offense': 'Violent',
    'Sexual Assault': 'Violent',
    'Assault with Deadly Weapon': 'Violent',
    'Kidnapping': 'Violent',
    'Arson': 'Violent',
    'Robbery': 'Violent',
    
}

# Add a new column 'Crime Category' based on the mapping
df['Crime Category'] = df['parent_incident_type'].map(crime_category)

# Count the occurrences of each category
crime_category_counts = df['Crime Category'].value_counts()

# Create a bar plot
plt.figure(figsize=(8, 6))
crime_category_counts.plot(kind='bar', rot=0)
plt.title('Violent Crimes in Santa Clara County')
plt.xlabel('Crime Category')
plt.ylabel('Number of Incidents')
plt.show()
```

```python
violent_incident_types = ['Assault', 'Weapons Offense', 'Other Sexual Offense', 'Sexual Assault', 'Assault with Deadly Weapon', 'Kidnapping', 'Arson', 'Robbery']

# Filter the DataFrame to include only violent crimes
violent_data = df[df['parent_incident_type'].isin(violent_incident_types)]

# Extract the month from the 'incident_datetime' column
violent_data['Month'] = pd.to_datetime(violent_data['incident_datetime']).dt.month

# Count the occurrences of each month
violent_month_counts = violent_data['Month'].value_counts().sort_index()

# Create a bar plot
plt.figure(figsize=(10, 6))
violent_month_counts.plot(kind='bar', rot=0)
plt.title('Violent Crimes per Month in Santa Clara County')
plt.xlabel('Month')
plt.ylabel('Number of Incidents')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.show()
```
